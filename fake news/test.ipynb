```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {
    'article': [
        'Economy is growing strong under the current administration',
        'Celebrity X caught in scandal, claims are unverified',
        'New research shows coffee reduces risk of heart disease',
        'Aliens found living on Mars, says scientist',
        'Local elections show shift towards environmental policies',
        'Moon landing was faked, new evidence suggests',
        'Breakthrough in cancer research promises cure',
        'Government corruption exposed in leaked documents'
    ],
    'is_real': [True, True, True, False, True, False, True, False]
}

df = pd.DataFrame(data)

def plot_word_frequencies(df, real_or_fake=True):
    # Filter the DataFrame
    filtered_df = df[df['is_real'] == real_or_fake]
    
    # Concatenate all articles into one giant string
    text = ' '.join(filtered_df['article']).lower()
    
    # Split the text into words and count each word's occurrence
    words = text.split()
    word_counts = Counter(words)
    
    # Convert the counter object into a DataFrame
    word_freq_df = pd.DataFrame(word_counts.items(), columns=['word', 'count']).sort_values(by='count', ascending=False)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(x='count', y='word', data=word_freq_df.head(10))
    title = 'Top 10 Words in Real News' if real_or_fake else 'Top 10 Words in Fake News'
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('Word')
    plt.show()

# Plot for real news
plot_word_frequencies(df, real_or_fake=True)

# Plot for fake news
plot_word_frequencies(df, real_or_fake=False)
```
